```python
# Togo Solar Data - EDA

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Display plots inline in Jupyter
%matplotlib inline

# Load Togo solar data
df = pd.read_csv('../data/togo-dapaong_qc.csv')

# Add Month column for further analysis
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Month'] = df['Timestamp'].dt.month

# Basic info and summary
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
print("\nMissing Values per Column:")
print(df.isna().sum())

# Handle missing values: numerical -> median, categorical -> mode
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].median(), inplace=True)
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Outlier detection for GHI
df['GHI_z'] = zscore(df['GHI'])
outliers = df[df['GHI_z'].abs() > 3]
print(f"\nPotential GHI outliers: {len(outliers)}")
print(outliers.head())

# Planned Visualizations
plt.figure(figsize=(12, 4))
plt.plot(df['Timestamp'], df['GHI'], color='orange')
plt.title("GHI over Time")
plt.xlabel("Timestamp")
plt.ylabel("GHI")
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='RH', y='GHI')
plt.title("RH vs GHI")
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(df['Tamb'], df['GHI'], s=df['RH'], alpha=0.5)
plt.title("GHI vs Tamb (Bubble size = RH)")
plt.xlabel("Tamb")
plt.ylabel("GHI")
plt.show()

plt.figure(figsize=(10, 8))
sns.heatmap(df[['GHI','DNI','DHI','TModA','TModB']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")# Togo Solar Data - EDA
# ==========================================
# This notebook performs exploratory data analysis (EDA) on Togo solar farm data.
# Tasks: load data, inspect, clean missing values, detect outliers, visualize relationships, and save cleaned CSV.

# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore

# Display plots inline
%matplotlib inline
sns.set(style="whitegrid")

# Load Togo solar data
df = pd.read_csv('../data/togo-dapaong_qc.csv', parse_dates=['Timestamp'])

# -------------------------------
# 1. Inspect Data
# -------------------------------
print("Dataset Info:\n")
print(df.info())
print("\nFirst 5 rows:\n", df.head())
print("\nSummary statistics:\n", df.describe())
print("\nMissing values per column:\n", df.isna().sum())

# -------------------------------
# 2. Handle Missing Values
# -------------------------------
# Fill numeric missing values with median
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# -------------------------------
# 3. Detect Outliers (GHI)
# -------------------------------
df['GHI_z'] = zscore(df['GHI'])
outliers = df[df['GHI_z'].abs() > 3]
print(f"\nPotential GHI outliers: {len(outliers)}")
print(outliers.head())

# -------------------------------
# 4. Planned Visualizations
# -------------------------------

# Line plot: GHI over time
plt.figure(figsize=(12,4))
plt.plot(df['Timestamp'], df['GHI'], color='orange', alpha=0.7)
plt.title('GHI over Time')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.show()

# Scatter plot: RH vs GHI
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='RH', y='GHI', alpha=0.5)
plt.title('Relative Humidity vs GHI')
plt.show()

# Bubble chart: GHI vs Tamb (size = RH)
plt.figure(figsize=(8,5))
plt.scatter(df['Tamb'], df['GHI'], s=df['RH'], alpha=0.3, c='skyblue', edgecolor='k')
plt.xlabel('Tamb')
plt.ylabel('GHI')
plt.title('GHI vs Tamb (Bubble size = RH)')
plt.show()

# Heatmap: correlation matrix
plt.figure(figsize=(10,8))
sns.heatmap(df[['GHI','DNI','DHI','TModA','TModB']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# -------------------------------
# 5. Save Cleaned Dataset
# -------------------------------
df.to_csv('../data/togo_clean.csv', index=False)
print("\nCleaned dataset saved as 'togo_clean.csv'")

plt.show()

# Save cleaned dataset
df.to_csv('../data/togo_clean.csv', index=False)
print("\n Cleaned dataset saved as '../data/togo_clean.csv'")
```
